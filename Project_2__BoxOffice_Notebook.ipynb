## Task 3: Relationship between Film Revenue and Budget

```python
T=df[['id','original_title','budget','revenue']].sort_values(['revenue'], ascending=False).head(10)
T
```

```python
A=df['revenue']
B=df['budget']
```

```python
sns.scatterplot(x=A,y=B)
```

```python
sns.regplot(x=A,y=B)
```

```python
R=df[['revenue','budget']].corr()
print(R)
sns.heatmap(R)
```

**Ans: Revenue and Budget are positively correlated with corr=0.752965**
